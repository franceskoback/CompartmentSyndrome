```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_hours_to_procedure_histogram(data):
    # Define the column names for hours to procedure
    hour_columns = ['HOURTOPROC', 'HOURTOPRO', 'ProcedureMins', 'HOSPITALPROCEDURESTARTHRS']

    # Find the column that exists in the data and is not NaN
    valid_column = None
    for column in hour_columns:
        if column in data.columns and not pd.isnull(data[column]).all():
            valid_column = column
            break

    if valid_column is None:
        print("No valid column for hours to procedure found.")
        return

    # Convert ProcedureMins column to hours if needed
    if valid_column == 'ProcedureMins':
        data['HOURTOPROC'] = data[valid_column] / 60
        valid_column = 'HOURTOPROC'

    # Filter out rows with NaN values in the selected column
    data = data[~data[valid_column].isnull()]

    # Define the time intervals
    #intervals = [0, 6, 30, 72, 150, 250, 400, 700, 1200, 2500]
    intervals = [0, 6, 30, 2500]

    # Group the data based on the intervals
    grouped_data = data.groupby(pd.cut(data[valid_column], intervals)).size()

    # Create the histogram
    plt.bar(grouped_data.index.astype(str), grouped_data.values)

    # Set the labels and title
    plt.xlabel('Time Intervals')
    plt.ylabel('Frequency')
    plt.title('Histogram of Hours to Procedure')

    # Rotate the x-axis labels for better visibility
    plt.xticks(rotation=45)

    # Show the histogram
    plt.show()
```

```python
def calculate_demographic_ratio_diff(combined_df, demographic_factor, group_0_6, group_30_plus):
    # Calculate ratios for different demographic groups in the 0-6 hours group
    group_0_6_demo_ratio = len(group_0_6[(group_0_6[demographic_factor] == 1)]) / len(group_0_6)

    # Calculate ratios for different demographic groups in the 6-30 hours group
    #group_6_30_demo_ratio = len(group_6_30[(group_6_30[demographic_factor] == 1)]) / len(group_6_30)

    # Calculate ratios for different demographic groups in the 30+ hours group
    group_30_plus_demo_ratio = len(group_30_plus[(group_30_plus[demographic_factor] == 1)]) / len(group_30_plus)

    # Calculate the differences in ratios between the 30+ hours and 0-6 hours groups
    demo_ratio_diff = group_30_plus_demo_ratio - group_0_6_demo_ratio

    return demo_ratio_diff
```
